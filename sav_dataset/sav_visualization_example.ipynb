```python
from utils.sav_utils import DAVISDataset
# Import the adapter
from utils.davis_adapter import DAVIStoSAVAdapter

try:
    from utils.sav_utils import DAVISDataset
except ImportError:
    print("utils.sav_utils module not found, using local davis_utils instead")
    # Option 2: If you saved it as davis_utils.py in the current directory 
    from davis_utils import DAVISDataset

# Initialize the dataset with your DAVIS path
davis_dir = r"C:\Users\fixyo\Downloads\DAVIS-2017-trainval-480p\DAVIS"
sav_dataset = DAVISDataset(davis_dir=davis_dir)

# Print available sequences
print("Available sequences in DAVIS dataset:")
sequences = sav_dataset.list_available_sequences()
print(f"First 10 sequences: {sequences[:10]}")

# Load a sequence
sequence_name = "walking"  # Try with your sequence
print(f"\nLoading sequence: {sequence_name}")
frames, manual_annot, auto_annot = sav_dataset.get_frames_and_annotations(sequence_name)

# Check if frames loaded successfully
if frames is not None and len(frames) > 0:
    print(f"Successfully loaded {len(frames)} frames")
    
    if manual_annot is not None:
        print(f"Successfully loaded annotations for {len(manual_annot['masklet'])} frames")
        
        # Check if we have annotations for each frame
        for i, frame_annot in enumerate(manual_annot['masklet']):
            print(f"Frame {i}: {len(frame_annot)} annotation objects")
        
        # Visualize multiple frames to verify different annotations
        print("\nVisualizing sequence with multiple frames to verify annotation alignment:")
        sav_dataset.visualize_sequence(
            sequence_name, 
            frame_indices=[0, 5, 10, 15, 20] if len(frames) > 20 else None
        )
    else:
        print("No annotations were loaded")
else:
    print(f"Failed to load frames for '{sequence_name}'")
    
    # Try another sequence as fallback
    fallback_sequence = sequences[0] if sequences else None
    if fallback_sequence:
        print(f"\nTrying fallback sequence '{fallback_sequence}':")
        sav_dataset.visualize_sequence(fallback_sequence)
```
